## Estructuras de control

Imaginemos que trabajamos en el área de finanzas o ventas de una empresa multinacional. Te entregan un archivo en Excel con 10,000 transacciones y te piden dos cosas urgentes:
1. Aplicar un 15% de descuento **SOLO** a las compras que superen los $500 dólares.
2. Enviar un correo automatizado a cada uno de esos 10,000 clientes.

Hacer esto manualmente te tomaría días enteros y el margen de error humano sería altísimo. Aquí es donde la programación brilla. Hoy aprenderemos a darle "inteligencia" y "velocidad" a nuestro código a través de las **Estructuras de Control**:
* **Condicionales (`if`, `elif`, `else`):** Para que Python tome decisiones por ti ("Si pasa esto, haz aquello").
* **Ciclos (`for` y `while`):** Para que Python repita una tarea miles de veces en una fracción de segundo.


### Condicionales
Le indicamos a python como tomar decisiones basadas en condiciones

```python
# Imagina que evaluamos el monto de compra de un cliente
monto_compra = 320

# Usamos 'if' (Si), 'elif' (Sino, si) y 'else' (Sino)
if monto_compra >= 500:
    print("¡Cliente VIP! Aplicar 15% de descuento")
    
elif monto_compra >= 200:
    print("Cliente Frecuente. Aplicar 5% de descuento")
    
else:
    print("Cliente Regular. No aplica descuento en esta compra")
```

### Ciclo FOR: Repeticiones aplicadas a una secuencia de objetos
Un ciclo `for` es como una línea de ensamblaje en una fábrica. Le damos a Python una colección de datos (como una lista de ventas diarias) y le pedimos que ejecute una acción **por cada uno de los elementos**, uno tras otro, hasta terminar.

```python
# Tenemos una lista con las ventas de la semana
listado_de_ventas = [2000,1000,6000,7500,3800,9200,1000]
total_ventas = 0

# Le decimos a Python: "Por cada venta en las ventas de la semana, suma el valor al total"
for x in listado_de_ventas:
    total_ventas = total_ventas + x
    print(f"Procesando una venta de: ${x}, llevamos un acumulado de ${total_ventas}")
    

print("El total de ingresos de la semana es: $", total_ventas)
```

```python
# Tenemos una lista con las ventas de la semana

venta_lunes = 2000
venta_martes = 1000
venta_miercoles = 6000
venta_jueves = 7500
venta_viernes = 3800
venta_sabado = 9200
venta_domingo = 1000

listado_de_ventas = [venta_lunes, venta_martes, venta_miercoles, venta_jueves, venta_viernes, venta_sabado, venta_domingo]
total_ventas = 0

### Palabra clave continue

for x in listado_de_ventas:
    if x < 2000:
        continue
    total_ventas = total_ventas + x
    print(f"Procesando una venta de: ${x}, llevamos un acumulado de ${total_ventas}")
    

print("El total de ingresos de la semana es: $", total_ventas)
```

```python
# Tenemos una lista con las ventas de la semana

venta_lunes = 2000
venta_martes = 1000
venta_miercoles = 6000
venta_jueves = 7500
venta_viernes = 3800
venta_sabado = 9200
venta_domingo = 1000

listado_de_ventas = [venta_lunes, venta_martes, venta_miercoles, venta_jueves, venta_viernes, venta_sabado, venta_domingo]
total_ventas = 0

### Palabra clave break

for x in listado_de_ventas:
    if x < 2000:
        continue
    
    total_ventas = total_ventas + x
    print(f"Procesando una venta de: ${x}, llevamos un acumulado de ${total_ventas}")
    
    if x > 7000:
        break 
    

print("El total de ingresos de la semana es: $", total_ventas)
```

### Ciclo WHILE: Repetir hasta cumplir una meta

A diferencia del ciclo `for` (que sabe exactamente cuántos elementos hay en una lista), el ciclo `while` se repite una y otra vez **mientras una condición siga siendo verdadera (True)**. 

Piensa en el `while` como el funcionamiento de un auto: "Mientras el auto esté acelerando, sigue bombeando gasolina".

```python
# Imaginemos que nustro objetivo es lograr $1000 en ventas.
meta_ventas = 1000
ventas_acumuladas = 0
numero_de_clientes = 0

# Mientras las ventas acumuladas sean menores a la meta:
while ventas_acumuladas < meta_ventas:
    # Simulamos que cada cliente compra exactamente $250
    ventas_acumuladas = ventas_acumuladas + 250
    numero_de_clientes = numero_de_clientes + 1
    print("Venta realizada. Acumulado actual: $", ventas_acumuladas)

print("¡Meta de $1000 alcanzada!")
print("Necesitamos", numero_de_clientes, "clientes para lograr la meta.")
```

```python
### Ciclo While para controlar la ejecución de un programa

opcion = 0
saldo = 1000
while opcion != 4 and saldo > 0:
    print("Bienvenido a Banco de la risa:")
    print("1. Retirar dinero")
    print("2. Consignar dinero")
    print("3. Consultar saldo")
    print("4. Salir")
    
    opcion = int(input("Seleccione una opción: "))
    
    if opcion == 1:
        print("Has seleccionado la opción de retiro.")
        retiro = int(input("Ingrese el monto a retirar: "))
        
        if retiro < 0:
            print("No se permite retirar montos negativos.")
            
        while retiro < 0:
            print("Monto inválido. Por favor ingrese un monto positivo.")
            retiro = int(input("Ingrese el monto a retirar: "))
            
        print(f"Has retirado ${retiro}")
        saldo -= retiro
        
    elif opcion == 2:
        print("Has seleccionado la opción de consignación.")
        consignacion = int(input("Ingrese el monto a consignar: "))
        print(f"Has consignado ${consignacion}")
        saldo += consignacion
    elif opcion == 3:
        print("Has seleccionado la opción de consulta de saldo.")
        print(f"Tu saldo actual es de ${saldo}")
    elif opcion == 4:
        print("Saliendo del programa...")
    else:
        print("Opción no válida, por favor intente nuevamente.")
        break
```
