# Tipos de datos - Errores

## Conversion de tipos de datos

### Enteros

```python
numero_string = "13" # Cadena de caracteres
numero_entero = int(numero_string)
print(numero_entero * 1)
print(f"Tipo de dato de la variable numero_string: {type(numero_string)}    str = cadena de caracteres, texto")
print(f"Tipo de dato de la variable numero_entero: {type(numero_entero)}    int = integer, numero entero")
```

### Decimales

```python
temperatura_decimal = 28.6
temperatura_entero = int(temperatura_decimal)
print(temperatura_decimal,temperatura_entero)
print(f"Tipo de dato de la variable temperatura_decimal: {type(temperatura_decimal)}    float = numero con decimales")
print(f"Tipo de dato de la variable temperatura_entero: {type(temperatura_entero)}")
```

```python
edad = 25 
edad_decimales = float(edad)
edad_duplicada = edad_decimales * 2
print(edad, edad_decimales, edad_duplicada)
```

```python
edad_numerica = 25
edad_string = str(edad)
print(edad_numerica, "mi edad es:"+ edad_string)
```

### Redondeo o aproximación de decimales a enteros

```python
# Ejemplo de aproximación antes de convertir a int
precio_original = 99.9  # Precio con decimales
precio_redondeado = round(precio_original)  # Redondear al valor entero más cercano
precio_entero = int(precio_original)  # Convertir a entero
print(f"Precio redondeado: {precio_redondeado}")
print(f"Precio entero: {precio_entero}")
```

### Cadenas de caracteres

```python
edad = 25 # Edad como numero entero
edad_string = str(edad) # Edad convertida a cadena de caracteres
print("Mi edad es " + edad_string) #concatenacion
```

```python
temperatura = 25.5 # Temperatura en grados
temperatura_string = str(temperatura) # Convertirlo a cadena
print("La temperatura actual es: " + temperatura_string + " grados Celsius")
```

# Manejo de errores

```python
numero = input("Ingrese un número: ")
numero_entero = int(numero) # Convertir la entrada a entero
print(f"El número ingresado es: {numero_entero} y su tipo de dato es: {type(numero_entero)}")
```

```python
try:
    numero = input("Ingrese un número: ")

    if "." in numero:
        raise ValueError("No puede ingresar numeros decimales")
    elif numero[0] == "-":
        if numero[1:].isdigit():
            numero = int(numero)
            raise ValueError("No puede ingresar numeros negativos")
        else:
            raise ValueError("No es un numero valido")
    else:
        if numero.isdigit():
            numero = int(numero)
        else:
            raise ValueError("No es un numero valido")

    if numero == 0:
        raise ValueError("No puede ingresar el numero 0")

    resultado = numero
    print("La edad del usuario es:", resultado)
except Exception as e:
    print(e)
```

```python
# Manejo de errores con try except
try:
    numero = int(input("Ingrese un número: "))
    resultado = 10 / numero
    print("El resultado es:", resultado)
except ZeroDivisionError:
    print("Error: No se puede dividir por cero.")
except ValueError:
    print("Error: Ingrese un numero valido")
except TypeError:
    print("Error: El tipo de dato es incorrecto")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Intentamos dividir un número entre cero
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
# Intentamos hacer operaciones con variables mal definidas
try:
    a = 10
    b = "texto"
    result = a + b  # Esto causará un TypeError
except TypeError as e:
    print(f"Error: No puedes sumar numeros y texto")
```

```python
# Intentamos abrir un archivo  #READ
# Archivo no encontrado: FileNotFoundError
try:
    with open("archivo_inexistente.txt", "r") as file:
        content = file.read()
except FileNotFoundError as e:
    print(f"Error: {e}")
finally:
    print("Este bloque siempre se ejecuta, aunque haya un error.")
```

```python
# Intentamos convertir una cadena no numérica en un número
try:
    num = int("texto")
except Exception as e:
    print(f"Se produjo un error genérico: {e}")
```

```python
# Intentamos acceder a un índice fuera de rango
my_list = [10, 20, 30]
try:
    print(my_list[5])
except IndexError as e:
    print(f"Error de índice: {e}")
```

```python
# Intentamos acceder a una clave que no existe
my_dict = {"nombre": "Juan", "edad": 30}
try:
    print(my_dict["direccion"])
except KeyError as e:
    print(f"Error de clave: {e}")
```

## Ejemplo Cajero Automatico

```python
saldo_inicial = 55000

retiro = float(input("Monto a retirar: "))

saldo_inicial -= retiro
print(f"Retiro exitoso. Saldo restante: ${saldo_inicial}")
```

```python
### Que el usuario ingrese la clave y el usuario correcto
### Si tenga la plata suficiente para retirar, sino que muestre un mensaje de error
### Retirar solo multiplos de 10.000
saldo = 45000000

usuario_base = "abril1234"
contraseña_base = "Abril202603"

try:
  usuario = input("Ingrese su usuario: ")
  if usuario != usuario_base: 
    raise ValueError("Usuario incorrecto")
  else:
    contraseña = input("Ingrese su contraseña: ")
    if contraseña != contraseña_base:
      raise ValueError("Contraseña incorrecta")
    else:
      retiro = float(input("Monto a retirar: "))
      if retiro <= 0:
        raise ValueError("El monto a retirar debe ser mayor a cero")
      elif retiro % 10000 != 0:
        raise ValueError("El monto a retirar debe ser un múltiplo de 10.000")
      elif retiro > saldo:
        raise ValueError("Saldo insuficiente para realizar el retiro")
      elif retiro < saldo:
        gravamen = retiro * 4 / 1000
        if retiro + gravamen > saldo:
          raise ValueError("Saldo insuficiente para cubrir el retiro y el impuesto gravamen")
        else:
          #saldo -= (retiro + gravamen)
          saldo = saldo - (retiro + gravamen)
          print(f"Retiro exitoso. Saldo restante: ${saldo}")   
except Exception as e:
  print(e)
  
  
```
