# Definición de Funciones
Piensa en una función como una máquina especializada en una fábrica:
1. Le entregas materia prima (tus datos).
2. La máquina hace un proceso interno que tú definiste previamente.
3. Te devuelve un producto terminado limpio y estandarizado (tu resultado).

En Python, la regla de oro es: **"No te repitas"** (Don't Repeat Yourself). Para solucionar esto, utilizamos **Funciones**.

```python
# Usamos la palabra reservada 'def' (definir), el nombre de la función y paréntesis.

def calcular_salario_final(salario_base):
    # Todo lo que esté indentado (con un espacio a la izquierda) pertenece al proceso interno de la función
    impuesto = salario_base * 0.10
    bono_desempeno = 500
    salario_neto = (salario_base - impuesto) + bono_desempeno
    
    # 'return' es la orden de entregar el producto terminado hacia afuera
    return salario_neto
```

```python
# En lugar de hacer la matemática manual, solo pasamos los salarios de nuestros empleados:

empleado_1 = calcular_salario_final(2000)
empleado_2 = calcular_salario_final(3500)

print("El salario final del Empleado 1 es: $", empleado_1)
print("El salario final del Empleado 2 es: $", empleado_2)
```

### Las piezas clave de una Función

Toda función de Python tiene estos elementos:
1. **`def`**: Comando de python indicando que se está creando una nueva función.
2. **Nombre de la función**: Debe ser un verbo claro que describa su propósito (ej. `limpiar_texto` o `calcular_ventas`).
3. **Parámetros**: Las variables dentro de los paréntesis `(salario_base)` son los datos que la función necesita para trabajar.
4. **Cuerpo de la función**: El bloque de código indentado que ejecuta la lógica.
5. **`return`**: La instrucción final que permite que la función te devuelva un valor para que lo sigas utilizando en el resto de tu análisis de datos.


```python
# Ejemplo con múltiples parámetros

def calcular_rentabilidad(ingresos, costos = 0):
    ganancia = ingresos - costos
    rentabilidad = (ganancia / costos) * 100
    return rentabilidad

# Llamamos a la máquina entregando ambas materias primas
proyecto_a = calcular_rentabilidad(ingresos=50000, costos=30000)

# También podemos pasar los valores directamente en el mismo orden
proyecto_b = calcular_rentabilidad(120000, 100000) 

print("Rentabilidad Proyecto A:", proyecto_a, "%")
print("Rentabilidad Proyecto B:", proyecto_b, "%")
```

# Manejo de errores

```python
numero = input("Ingrese un número: ")
numero_entero = int(numero) # Convertir la entrada a entero
print(f"El número ingresado es: {numero_entero} y su tipo de dato es: {type(numero_entero)}")
```

```python
try:
    numero = input("Ingrese un número: ")

    if "." in numero:
        raise ValueError("No puede ingresar numeros decimales")
    elif numero[0] == "-":
        if numero[1:].isdigit():
            numero = int(numero)
            raise ValueError("No puede ingresar numeros negativos")
        else:
            raise ValueError("No es un numero valido")
    else:
        if numero.isdigit():
            numero = int(numero)
        else:
            raise ValueError("No es un numero valido")

    if numero == 0:
        raise ValueError("No puede ingresar el numero 0")

    resultado = numero
    print("La edad del usuario es:", resultado)
except Exception as e:
    print(e)
```

```python
# Manejo de errores con try except
try:
    numero = int(input("Ingrese un número: "))
    resultado = 10 / numero
    print("El resultado es:", resultado)
except ZeroDivisionError:
    print("Error: No se puede dividir por cero.")
except ValueError:
    print("Error: Ingrese un numero valido")
except TypeError:
    print("Error: El tipo de dato es incorrecto")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Intentamos dividir un número entre cero
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
# Intentamos hacer operaciones con variables mal definidas
try:
    a = 10
    b = "texto"
    result = a + b  # Esto causará un TypeError
except TypeError as e:
    print(f"Error: No puedes sumar numeros y texto")
```

```python
# Intentamos abrir un archivo  #READ
# Archivo no encontrado: FileNotFoundError
try:
    with open("archivo_inexistente.txt", "r") as file:
        content = file.read()
except FileNotFoundError as e:
    print(f"Error: {e}")
finally:
    print("Este bloque siempre se ejecuta, aunque haya un error.")
```

```python
# Intentamos convertir una cadena no numérica en un número
try:
    num = int("texto")
except Exception as e:
    print(f"Se produjo un error genérico: {e}")
```

```python
# Intentamos acceder a un índice fuera de rango
my_list = [10, 20, 30]
try:
    print(my_list[5])
except IndexError as e:
    print(f"Error de índice: {e}")
```

```python
# Intentamos acceder a una clave que no existe
my_dict = {"nombre": "Juan", "edad": 30}
try:
    print(my_dict["direccion"])
except KeyError as e:
    print(f"Error de clave: {e}")
```
