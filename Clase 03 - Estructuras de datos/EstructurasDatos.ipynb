### Tuplas (Inmutable - Longitud Fija)

````python
# Crear la tupla a partir de una secuencia de valores entre parentesis separados por comas
tupla = (4, 5, 6)
print(tupla)
````

````python
# Crear la tupla a partir de una secuencia de valores separados por comas
tupla_2 = 4, 5, 6
print(tupla_2)
print(type(tupla_2))
````

````python
# Crear la tupla usando la funcion tuple sobre cualquier secuencia de datos
tupla_3 = tuple("hola")
tupla_3
````

````python
# Crear la tupla usando la funcion tuple sobre cualquier secuencia de datos
tupla_4 = tuple([10,20,5555])
tupla_4
````

````python
#Revisar si un objeto pertenece a la tupla
print(f"El número 10 está en la tupla_4? {10 in tupla_4}")
print(f"El número 50 está en la tupla_4? {50 in tupla_4}")
````

````python
#Acceder a los datos de la tupla
print(f"Primer elemento de la tupla: {tupla_4[0]}") #INDEXACION
print(f"Segundo elemento de la tupla: {tupla_4[1]}")
print(f"Tercer elemento de la tupla: {tupla_4[2]}")
````

````python
tupla_5 = tuple(range(0,41,5))
tupla_5
````

````python
#Las tuplas no se pueden modificar
tupla_5[10] = 11  
````

### Listas (Modificable - Longitud Variable)

````python
lista = [2, 3, 7, None]
lista
````

````python
lista_2 = list(range(20,200,20))
lista_2
````

````python
#Las listas se pueden modificar
lista_2[3] = 5000000
lista_2
````

````python
#Puedo agregar objetos nuevos al final de la lista con append()
lista_2.append(80)
lista_2
````

````python
#Puedo remover objetos de la lista con pop()
lista_2.pop(3) # Entre parentesis se especifica la posición del objeto que quiero remover
lista_2
````

````python
lista_2.pop() # si no se especifica ningun valor se remueve el ultimo objeto de la lista
lista_2
````

````python
lista_2.pop(-4) # Si se especifica un valor negativo se cuenta desde el final de la lista
lista_2
````

````python
#Revisar si un objeto pertenece a la lista
print(f"El número 20 está en la lista_2? {20 in lista_2}")
print(f"El número 50 está en la lista_2? {50 in lista_2}")
````

### Diccionarios (Llave - Valor)

````python
[datos_personales,datos_personales,datos_personales,datos_personales] ## NO SE DEBE HACER
````

````python
datos_personales = {"id":1231412312,
                    "nombre": "Cristian",
                    "edad": 10,
                    "equipo_favorito":"colombia",
                    "hijos":[{
                              "nombre":"Sofia",
                              "edad": 5
                            },
                            {
                              "nombre":"Santiago",
                              "edad": 3
                            }
                  ]
                  }
datos_personales
````

````python
# Puedo leer los datos almacenados en una llave usando parentesis cuadrados
datos_personales["equipo_favorito"]
````

````python
# Tambien puedo leer los datos almacenados en una llave usando la funcion get()
datos_personales.get("equipo_favorito")
````

````python
# Para agregar nuevas llaves al diccionario, declaro la nueva llave entre parentesis cuadrados y le asigno un nuevo valor
datos_personales["fecha_nacimiento"] = "31/08/1998"
datos_personales["signozodiacal"] = "Sagitario"

datos_personales
````

````python
# Puedo borrar datos del diccionario usando la sentencia 'del' delete
del datos_personales['signozodiacal']
datos_personales
````

````python
# Tambien puedo borrar datos del diccionario usando la funcion pop() especificando la llave que quiero eliminar
datos_personales.pop("fecha_nacimiento")
datos_personales
````

````python
## Modificar los datos de una llave 
datos_personales["edad"] = 27
datos_personales
````
