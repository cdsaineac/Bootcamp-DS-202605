## Estructuras de datos



### Tuplas (Inmutable - Longitud Fija)

````python
# Crear la tupla a partir de una secuencia de valores entre parentesis separados por comas
estudiantes = ('Ana', 'Cristian','Gabriel','Helen')
print(estudiantes)
````

````python
#Revisar si un objeto pertenece a la tupla
print(f"Cristian está en la tupla estudiantes? {'Cristian' in estudiantes}")
print(f"El número 50 está en la tupla estudiantes? {50 in estudiantes}")
````

````python
#Acceder a los datos de la tupla
print(f"Primer elemento de la tupla: {estudiantes[0]}") #INDEXACION
print(f"Segundo elemento de la tupla: {estudiantes[1]}")
print(f"Tercer elemento de la tupla: {estudiantes[2]}")
````

### Listas (Modificable - Longitud Variable)

````python
estudiantes = ['Ana', 'Cristian', 'Gabriel', 'Helen']
estudiantes
````

````python
estudiantes.append("Pablo")
````

````python
estudiantes
````

````python
#Las listas se pueden modificar
estudiantes[0] = "Ana María"
estudiantes[2] = "Enrique"
````

````python
estudiantes
````

````python
#Puedo remover objetos de la lista con pop()
estudiantes.pop(2) # Entre parentesis se especifica la posición del objeto que quiero remover
````

````python
estudiantes.pop() # si no se especifica ningun valor se remueve el ultimo objeto de la lista
````

````python
estudiantes
````

````python
# Entre parentesis se especifica el objeto que quiero remover
estudiantes.remove("Helen")
````

````python
estudiantes.index("Helen") # Si el objeto no se encuentra en la lista se genera un error
````

````python
estudiantes.append('Helen')
````

````python
estudiantes
````

````python
#Revisar si un objeto pertenece a la lista
print(f"El estudiante Helen está en la lista de estudiantes? {"Helen" in estudiantes}")
print(f"El número 50 está en la lista de estudiantes? {50 in estudiantes}")
````

### Diccionarios (Llave - Valor)

````python
estudiantes = {
    165465165:{
        "nombre":"Cristian García",
        "edad": 30,
        "cursos": ["Python", "SQL", "Power BI"]
    },
    165465166:{
        "nombre":"Cristian García",
        "edad": 30,
        "cursos": ["Python", "SQL", "Power BI"]
    },
    165465167:{
        "nombre":"Gabriel García",
        "edad": 25,
        "cursos": ["Java", "C++", "JavaScript"]
    },
    165465168:{
        "nombre":"Helen Martinez",
        "edad": 28,
        "cursos": ["R", "SAS", "SPSS"]
    }
}
````

````python
estudiantes[165465166]
````

````python
estudiantes.get(165465165).get("cursos")
````

````python
estudiantes
````

````python
estudiantes[165465169] = {
    "nombre":"Sebastian Perez",
    "edad": 35,
    "cursos": ["Python", "SQL", "Power BI"]
}
````

````python
# Para agregar nuevas llaves al diccionario, declaro la nueva llave entre parentesis cuadrados y le asigno un nuevo valor
estudiantes[165465165]["edad"] = 31
estudiantes[165465165]["correo"] = "cristian.garcia@example.com"
estudiantes[165465165]
````

````python
# Puedo borrar datos del diccionario usando la sentencia 'del' delete
del estudiantes[165465169]
estudiantes
````

````python
# Tambien puedo borrar datos del diccionario usando la funcion pop() especificando la llave que quiero eliminar
estudiantes.pop(165465165)
estudiantes
````
